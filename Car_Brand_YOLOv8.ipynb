# Car Brands Predictor- YOLOv8

In this project will be designed a model capable to recognize brands of cars. Now the difference is the model will be YOLOv8.

# Importations

```python
import os
from ultralytics import YOLO 
```

# Model Configuration

```python
path = 'LOCAL_DIR'
```

The version chosen is "yolov8n-cls", the smaller version of yolov8, in the function of classifier. The meanly reason are the computational power needed by the model.

```python
model = YOLO("yolov8n-cls.pt")

results = model.train(
    data= path, 
    epochs=10,  
    batch=8,
    cache='disk',
    device=0,
    workers=4
    )
```

# Result

**Top1_acc = 88,9 %**    
**Top5_acc = 97,9 %**

 Top1_acc accuracy measures the proportion of predictions where the model's top-ranked (i.e., most confident) prediction matches the true label of the data sample. In other words, it evaluates whether the model's highest-confidence prediction is correct.


 Top5_acc is a more lenient metric compared to Top-1 accuracy. Instead of considering only the top-ranked prediction, it evaluates whether the true label is among the top 5 predictions made by the model. This is useful for cases where the classes are closely related or ambiguous. 

The result obtained with this model are higher, is easier to train than transfer learning process, but needs to much compute power.
